## Carregamento dos dados

```python
# importe os dados do dataset em datasets/abandonos-202409131927.csv

import pandas as pd
import numpy as np

anoletivo = 2021
df = pd.read_parquet(f'datasets/boletim-abandonos.parquet/nr_anoletivo={anoletivo}')
df['nr_anoletivo'] = anoletivo

df['areatrabalho'] = np.where(df['ci_areatrabalho'] == 14, 'pt', 'mat')
df.drop(columns='ci_areatrabalho', inplace=True)

print(df.columns)
df.head()
```

## Transformando registros em features

```python
columns_id = ['ci_aluno', 'fl_sexo', 'aluno_municipio_id', 'aluno_estado',
       'aluno_municipio_latitude', 'aluno_municipio_longitude', 'ci_turma',
       'cd_etapa', 'cd_turno', 'ci_unidade_trabalho',
       'unidade_trabalho_latitude', 'unidade_trabalho_longitude',
       'abandonou', 'nr_anoletivo', 'areatrabalho'
       ]

columns_id_sem_ci_areatrabalho = ['ci_aluno', 'fl_sexo', 'aluno_municipio_id', 'aluno_estado',
       'aluno_municipio_latitude', 'aluno_municipio_longitude', 'ci_turma',
       'cd_etapa', 'cd_turno', 'ci_unidade_trabalho',
       'unidade_trabalho_latitude', 'unidade_trabalho_longitude',
       'abandonou', 'nr_anoletivo'
       ]


df_group = df.melt(id_vars=columns_id, var_name='tipo_column', value_name='value')


# # # Criar o nome da nova coluna concatenando 'tipo_column' e 'disciplina_id'
df_group['new_col'] = df_group['tipo_column'] + '_' + df_group['areatrabalho'].astype(str)


# # drop columns ci_areatrabalho, tipo_column
df_group = df_group.pivot(index=columns_id_sem_ci_areatrabalho, columns='new_col', values='value').reset_index()

# # # Remover multiindex das colunas resultantes
df_group.columns.name = None

del columns_id
del columns_id_sem_ci_areatrabalho

df_group.head()
```

```python
#  gravar o dataset em datasets/boletim-abandonos-2024.parquet particionado por nr_anoletivo
df_group.to_parquet('datasets/dados-com-features.parquet', partition_cols=['nr_anoletivo'])
```
