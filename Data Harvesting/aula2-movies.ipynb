```python
import requests

url = 'https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
res = requests.get(url, headers=headers)
print(f'STATUS CODE: {res.status_code}')
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.content, 'html.parser')
title = soup.title.string
print(f'Titulo da página: {title}')
```

```python
# selecionando os nomes dos produtos e preço
top_items = []
products = soup.select('li.ipc-metadata-list-summary-item')
for i, product in enumerate(products):
    image = product.select('img')[0].get('src')
    name = product.select('h3.ipc-title__text')[0].text
    
    # pegue o rating do filme
    # rate esteja disponível na seguinte estrutura
    # <span aria-label="IMDb rating: 6.4" class="ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating" data-testid="ratingGroup--imdb-rating"><svg width="24" height="24" xmlns="http://www.w3.org/2000/svg" class="ipc-icon ipc-icon--star-inline" viewBox="0 0 24 24" fill="currentColor" role="presentation"><path d="M12 20.1l5.82 3.682c1.066.675 2.37-.322 2.09-1.584l-1.543-6.926 5.146-4.667c.94-.85.435-2.465-.799-2.567l-6.773-.602L13.29.89a1.38 1.38 0 0 0-2.581 0l-2.65 6.53-6.774.602C.052 8.126-.453 9.74.486 10.59l5.147 4.666-1.542 6.926c-.28 1.262 1.023 2.26 2.09 1.585L12 20.099z"></path></svg>6.4<span class="ipc-rating-star--voteCount">&nbsp;(31K)</span></span>
    # no caso acima o rating é 6.4

    rate = product.select('span.ipc-rating-star')[0].get('aria-label').replace('IMDb rating: ', '')

    top_items.append({
        'name': name,
        'rate': rate,
        'image': image
    })

top_items
```
