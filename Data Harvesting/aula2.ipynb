```python
import requests

url = 'https://codedamn-classrooms.github.io/webscraper-python-codedamn-classroom-website'
res = requests.get(url)

print(res.text)
print(f'STATUS CODE: {res.status_code}')
```

```python
from bs4 import BeautifulSoup

page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
title = soup.title.string
print(f'Titulo da página: {title}')
```

```python
# capiturando o head da página
head = soup.head
print(f'Head da página: {head}')
```

```python
# capturando o body da página
body = soup.body
print(f'Body da página: {body}')
```

```python
soup.select('h1')
```

```python
for i, h1 in enumerate(soup.select('h1')):
    print(f'{i} - {h1.text}')
```

```python
# selecionando os nomes dos produtos e preço
top_items = []
products = soup.select('div.thumbnail')
for i, product in enumerate(products):
    image = product.select('img')[0].get('src')
    name = product.select('a.title')[0].get('title')
    price = product.select('h4.price')[0].text
    reviews = product.select('div.ratings')[0].text.strip()
    top_items.append({
        'name': name,
        'price': price,
        'reviews': reviews,
        'image': image
    })

top_items
```

```python
images = soup.select('img')
for i, image in enumerate(images):
    print(f'{i} - {image.get("src")} - {image.get("alt")}')
```
