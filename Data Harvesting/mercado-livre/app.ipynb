```python
import requests

def get_json(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"URL: {url} Status: {response.status_code}")
```

```python
import pandas as pd
    
# BUSCA DE PRODUTOS
# https://api.mercadolibre.com/sites/MLB/search?q={QUERY}
def search_product(query):
    url = f"https://api.mercadolibre.com/sites/MLB/search?q={query}"
    data = get_json(url)
    return pd.json_normalize(data['results'])

df = search_product('processador Core I9')

df.head()

# selecione as coluns id, title, price e available_quantity
df = df[['id', 'title', 'price', 'available_quantity', 'category_id']]

# mostra os itens ordenados pelo  menor preço
df.sort_values('price').head()
```

```python
# Detalhes do produto
# https://api.mercadolibre.com/items/{ID}
def detail_product(id):
    url = f"https://api.mercadolibre.com/items/{id}"
    data = get_json(url)
    return pd.json_normalize(data)

# exibe os detalhes do produto com id MLB4693899210
detail_product('MLB4693899210')
```

```python
# CATEGORIA
# https://api.mercadolibre.com/categories/{CATEGORY_ID}
def category(id):
    url = f"https://api.mercadolibre.com/categories/{id}"
    data = get_json(url)
    return pd.json_normalize(data)

category('MLB1693')
```

```python
# LOCALIDADES
# https://api.mercadolibre.com/countries

def countries():
    url = f"https://api.mercadolibre.com/countries"
    data = get_json(url)
    return pd.json_normalize(data)

countries().head()
```

```python
# https://api.mercadolibre.com/states/{STATE_ID}

def states(id):
    url = f"https://api.mercadolibre.com/states/{id}"
    data = get_json(url)
    print(data)
    return pd.json_normalize(data)

states('BR-SP')
```

```python

# https://api.mercadolibre.com/cities/{CITY_ID}
def cities(id):
    url = f"https://api.mercadolibre.com/cities/{id}"
    data = get_json(url)
    return pd.json_normalize(data)

cities('BR-SP-44')

```
