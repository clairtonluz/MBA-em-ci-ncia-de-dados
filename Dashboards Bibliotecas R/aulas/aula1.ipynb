```python
import gdown
import os

file = 'https://drive.google.com/file/d/1-NIsWOWYCGxbsDXfbE8X-Z7BmKfv1ldv'
output = 'DESCRITORESMATH.csv'

if not os.path.exists(output):
    print(f"File {output} not found. Downloading...")
    fileId = file.split('/')[-1]
    url = 'https://drive.google.com/uc?id=' + fileId
    print(f"Downloading from {url} to {output}")
    gdown.download(url, output, quiet=False)
else:
    print(f"File {output} already exists. No download needed.")

```

```python
import pandas as pd

df=pd.read_csv('DESCRITORESMATH.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

z=df['D50'].values
plt.scatter(df['D16'], df['D19'], s=100, c=z, alpha=0.5, edgecolors='black', lw=0.2, cmap='BrBG')
plt.xlabel('D16', fontsize=12)
plt.ylabel('D19', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(alpha=0.3)
plt.colorbar(label='D50')
plt.show()

```
