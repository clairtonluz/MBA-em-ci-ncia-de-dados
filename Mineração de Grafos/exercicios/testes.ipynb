```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo simples
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (4, 5), (6, 7), (6, 8), (7, 8), (8, 9)])

# Detectando comunidades usando algoritmo de Louvain
from community import community_louvain
partition = community_louvain.best_partition(G)
print ("Comunidades detectadas: ", partition)

# Desenhando o grafo com cores de comunidade
pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))

# Mapeando as cores das comunidades
colors = [partition[node] for node in G.nodes()]
nx.draw(G, pos, node_color=colors, with_labels=True, cmap=plt.cm.Paired, node_size=300)

# Mostrando o gráfico
plt.title("Detecção de Comunidades com Algoritmo de Louvain")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

# Cria um grafo
G = nx.Graph()

# Adiciona nós com pesos diferentes
G.add_node('A', weight=1.5)
G.add_node('B', weight=2.0)
G.add_node('C', weight=3.0)
G.add_node('D', weight=4.0)

# Adiciona arestas
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'D')
G.add_edge('D', 'A')

# Obtém os pesos dos nós
weights = [G.nodes[node]['weight'] for node in G.nodes()]

# Define uma função de mapeamento de cores com base nos pesos
norm = Normalize(vmin=min(weights), vmax=max(weights))
cmap = plt.get_cmap('viridis')

# Cria um mapeador de cores
sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

# Desenha o grafo com nós coloridos
pos = nx.circular_layout(G)  # Define uma disposição circular para os nós
nx.draw_networkx_nodes(G, pos, node_color=[cmap(norm(w)) for w in weights], node_size=500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)

# Adiciona a barra de cores
cbar = plt.colorbar(sm, orientation='vertical')
cbar.set_label('Pesos')

plt.axis('off')
plt.show()
```
