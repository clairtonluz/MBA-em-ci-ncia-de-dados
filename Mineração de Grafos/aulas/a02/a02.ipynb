# 5. Caminho

Caminho é qualquer sequência de vértices consecutivos ligados por arestas.

![picture](https://drive.google.com/thumbnail?id=1-9L3HrDpME9z_uR1tYglLnGkCTLiDo8V&sz=w400)

Aqui, comprimento é o número de arestas atravessadas ao longo de um caminho.

## 5.1 Caminhos geodésicos

Caminho geodésico é o caminho mais curto entre dois vértices.

![picture](https://drive.google.com/thumbnail?id=1ZE80jhGOfpgSsDGgnOE6GRDpBdvFKQ94&sz=w400)

Nesse contexto:

* O comprimento de um caminho geodésico é chamado de distância geodésica ou distância mais curta.
* Diâmetro é o comprimento do maior caminho geodésico entre dois vértices quaisquer da rede.

# 7. Dígrafos

## 7.1 Criando um dígrafo

```python
DG = nx.DiGraph()

DG.add_node('A')
DG.add_node('B')
DG.add_node('C')
DG.add_node('D')
DG.add_node('E')
DG.add_node('F')
DG.add_node('G')
DG.add_node('H')

DG.add_edge('A', 'B')
DG.add_edge('B', 'A')
DG.add_edge('A', 'C')
DG.add_edge('C', 'D')
DG.add_edge('D', 'C')
DG.add_edge('A', 'E')
DG.add_edge('E', 'A')

DG.add_edge('F', 'G')
DG.add_edge('G', 'F')

# print(DG.nodes)
# print(DG.nodes.data())
# print(DG.edges)
# print(DG.edges.data())

pos = nx.random_layout(DG)
nx.draw_networkx(DG, pos = pos)

# # print(DG.degree['A'])
print(DG.in_degree['A'])
print(DG.out_degree['A'])
```

## 7.3 Componente fortemente conectado

```python
#print(nx.is_strongly_connected(DG))
print(nx.number_strongly_connected_components(DG))

#print(list(nx.strongly_connected_components(DG)))
# for c in sorted(nx.strongly_connected_components(DG), key=len, reverse=True):
#     print(len(c),c)

lscc = max(nx.strongly_connected_components(DG), key=len)
print(lscc)

SDG = DG.subgraph(lscc).copy()
print(SDG)

list_of_strongly_components = [DG.subgraph(scc).copy() for scc in nx.strongly_connected_components(DG)]

for SDG in list_of_strongly_components:
    nx.draw_networkx(SDG)
    plt.show()
```

# 8. Multigrafos e Multidígrafos

```python
MG = nx.MultiGraph()

MG.add_node('A')
MG.add_node('B')

MG.add_edge('A', 'B',peso=1.0,rad='0.1')
MG.add_edge('A', 'B',peso=2.0,rad='0.5')

print(MG.nodes)
#print(MG.nodes.data())
print(MG.edges)
#print(MG.edges.data())

def plot_network(G):
    fig,ax = plt.subplots()
    pos = nx.spring_layout(G)

    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_labels(G, pos)
    for edge in G.edges.data():
        nx.draw_networkx_edges(G,
                               pos,
                               edgelist=[(edge[0],edge[1])],
                               connectionstyle='arc3, rad = {}'.format(edge[2]['rad']))

plot_network(MG)
```

```python
MDG = nx.MultiDiGraph()

MDG.add_node('A')
MDG.add_node('B')

MDG.add_edge('A', 'B',peso=1.0,rad='0.1')
MDG.add_edge('A', 'B',peso=2.0,rad='-0.1')

print(MDG.nodes)
#print(MDG.nodes.data())
print(MDG.edges)
#print(MDG.edges.data())

def plot_network(G):
    fig,ax = plt.subplots()
    pos = nx.spring_layout(G)

    nx.draw_networkx_nodes(G, pos)
    nx.draw_networkx_labels(G, pos)
    for edge in G.edges.data():
        nx.draw_networkx_edges(G,
                               pos,
                               edgelist=[(edge[0],edge[1])],
                               connectionstyle='arc3, rad = {}'.format(edge[2]['rad']))

plot_network(MDG)
```
