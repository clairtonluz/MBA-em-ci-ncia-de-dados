# AUTO ARIMA 

```python
from pmdarima.arima import auto_arima
```

```python
modelo_auto=auto_arima(ser3,trace=True,stepwise=True, seasonal=False,
                     max_p=12,max_q=2,max_P=12, max_Q=1, star_p=0,
                     star_P=0,star_Q=0,m=12)
```

```python
modelo_auto.aic()
```

```python
ser
```

```python
resultado=modelo_auto.fit(ser3)
```

```python
predicao_auto=resultado.predict(n_periods=20)
```

```python
predicao_auto
```

```python
plt.plot(ser3)
```

```python
serprev=list(ser.values)+list(predicao_auto**3)
```

```python
ser
```

```python
predicao=pd.Series(data=serprev,index=pd.date_range(start="1949-01-01",freq="M",periods=len(serprev)))
```

```python
plt.figure(dpi=300)
plt.plot(range(len(ser)),ser,label='ORIGINAL',lw=4)
plt.plot(range(len(predicao)),predicao,label='PREDICAO')
plt.legend()
plt.show()
```

# Testando ARMA

```python
ARMA=sm.tsa.arima.ARIMA(ser3.values,order=(12,0,1))
resultado=ARMA.fit()
print(resultado.summary())
```

```python
predicao=resultado.predict(start=144,end=200)
```

```python
serprev=list(predicao**3)
```

```python
plt.figure(dpi=300)
plt.plot(range(1,len(ser)+1),ser,label='ORIGINAL')
plt.plot(range(len(ser)+1,len(ser)+len(serprev)+1),serprev,label='PREDICAO',lw=4)
plt.legend()
plt.ylim(0,750)
plt.show()
```

```python
ARIMA=sm.tsa.arima.ARIMA(ser3.values,order=(12,1,1))
resultado=ARIMA.fit()
print(resultado.summary())
```

```python
predicao=resultado.predict(start=144,end=500)
```

```python
serprev=list(predicao**3)
```

```python
plt.figure(dpi=300)
plt.plot(range(1,len(ser)+1),ser,label='ORIGINAL')
plt.plot(range(len(ser)+1,len(ser)+len(serprev)+1),serprev,label='PREDICAO',lw=4)
plt.title("TRANSFORMA CUBICA")
plt.legend()
plt.show()
```
