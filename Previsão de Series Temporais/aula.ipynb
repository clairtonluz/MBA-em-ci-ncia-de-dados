```python

import numpy as np
serie = pd.Series(np.random.randint(1, 10, 10),
                  index=pd.date_range(start='2021-01-01', end='2021-01-10', freq='D'))
```

```python
# plot o grafico de linha da serie
serie.plot()
```

```python
from statsmodels.tsa.stattools import kpss

def is_stationary(serie):
    analise = kpss(serie)
    print(analise)
    p_value = analise[1]
    valor_critico = analise[3]['5%']
    return p_value < valor_critico

is_stationary(serie)
```
